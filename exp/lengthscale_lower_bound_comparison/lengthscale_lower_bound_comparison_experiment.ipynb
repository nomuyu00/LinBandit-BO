# 長さスケール下限 比較実験

この実験では、以下の2つのLinBandit-BOアルゴリズムを比較します：

1. ベースライン: 非正規化入力・長さスケール下限なし（勾配ベース報酬, L下限付きスケーリング）
2. 長さスケール下限: 入力を[0,1]^dに正規化し、RBFの長さスケールに下限 l_min=0.05 を課す（勾配ベース報酬, L下限付きスケーリング）

各関数について20回の独立実行を行い、収束履歴・最終性能・報酬履歴・次元別要約を比較します。

```python
# ベースラインLinBandit-BO（L下限付きスケーリング）
class BaselineLinBanditBO:
    """
    ベースラインLinBandit-BO: 勾配の絶対値を報酬として使用し、推定L(=L_hat)でスケーリング（L_min下限を適用）
    """
    
    def __init__(self, objective_function, bounds, n_initial=5, n_max=100, 
                 coordinate_ratio=0.8, n_arms=None):
        self.objective_function = objective_function
        self.bounds = bounds.float()
        self.dim = bounds.shape[1]
        self.n_initial = n_initial
        self.n_max = n_max
        self.coordinate_ratio = coordinate_ratio
        
        # 0.5x arms設定
        self.n_arms = n_arms if n_arms is not None else max(1, self.dim // 2)
        
        # Linear Banditのパラメータ
        self.A = torch.eye(self.dim)
        self.b = torch.zeros(self.dim)
        
        # 初期点の生成
        self.X = torch.rand(n_initial, self.dim) * (bounds[1] - bounds[0]) + bounds[0]
        self.X = self.X.float()
        
        # 状態変数
        self.Y = None
        self.best_value = None
        self.best_point = None
        self.model = None
        self.eval_history = []
        self.theta_history = []
        self.scale_init = 1.0
        self.total_iterations = 0
        
        # 報酬履歴の記録
        self.reward_history = []
        self.direction_history = []
        
        # Lの推定と下限（スパイク抑制のため）
        self.L_min = 0.1
        self.L_hat = max(1.0, self.L_min)
        
    def update_model(self):
        """ガウス過程モデルの更新"""
        base = RBFKernel(ard_num_dims=self.X.shape[-1])
        kernel = ScaleKernel(base).to(self.X)
        self.model = SingleTaskGP(self.X, self.Y, covar_module=kernel)
        mll = ExactMarginalLogLikelihood(self.model.likelihood, self.model)
        fit_gpytorch_model(mll)
        
    def initialize(self):
        """初期化：初期点での評価とモデル構築"""
        y_val = self.objective_function(self.X)
        self.Y = y_val.unsqueeze(-1).float()
        
        # スケーリング係数の計算
        y_max, y_min = self.Y.max().item(), self.Y.min().item()
        self.scale_init = (y_max - y_min) if (y_max - y_min) != 0 else 1.0
        
        # モデルの初期化
        self.update_model()
        
        # 最良点の初期化
        post_mean = self.model.posterior(self.X).mean.squeeze(-1)
        bi = post_mean.argmin()
        self.best_value = post_mean[bi].item()
        self.best_point = self.X[bi]
        self.eval_history = [self.best_value] * self.n_initial
        
    def generate_arms(self):
        """アーム生成（0.5x arms）"""
        num_coord = int(self.coordinate_ratio * self.n_arms)
        num_coord = min(num_coord, self.dim)
        
        # ランダムに座標を選択
        idxs = np.random.choice(self.dim, num_coord, replace=False)
        
        # 座標方向の生成
        coords = []
        for i in idxs:
            e = torch.zeros(self.dim, device=self.X.device)
            e[i] = 1.0
            coords.append(e)
            
        coord_arms = torch.stack(coords, 0) if coords else torch.zeros(0, self.dim, device=self.X.device)
        
        # ランダム方向の生成
        num_rand = self.n_arms - num_coord
        rand_arms = torch.randn(num_rand, self.dim, device=self.X.device) if num_rand > 0 else torch.zeros(0, self.dim, device=self.X.device)
        
        if num_rand > 0:
            norms = rand_arms.norm(dim=1, keepdim=True)
            rand_arms = torch.where(norms > 1e-9, rand_arms / norms, 
                                   torch.randn_like(rand_arms) / (torch.randn_like(rand_arms).norm(dim=1,keepdim=True)+1e-9))
            
        return torch.cat([coord_arms, rand_arms], 0)
    
    def select_arm(self, arms_features):
        """Linear UCBによる方向選択"""
        # LinUCBパラメータ
        sigma = 1.0
        L = 1.0
        lambda_reg = 1.0
        delta = 0.1
        S = 1.0
        
        # 現在のパラメータ推定
        A_inv = torch.inverse(self.A)
        theta = A_inv @ self.b
        self.theta_history.append(theta.clone())
        
        # 信頼幅の計算
        current_round_t = max(1, self.total_iterations)
        log_term_numerator = max(1e-9, 1 + (current_round_t - 1) * L**2 / lambda_reg)
        beta_t = (sigma * math.sqrt(self.dim * math.log(log_term_numerator / delta)) + 
                  math.sqrt(lambda_reg) * S)
        
        # UCBスコアの計算
        ucb_scores = []
        for i in range(arms_features.shape[0]):
            x = arms_features[i].view(-1, 1)
            mean = (theta.view(1, -1) @ x).item()
            try:
                var = (x.t() @ A_inv @ x).item()
            except torch.linalg.LinAlgError:
                var = (x.t() @ torch.linalg.pinv(self.A) @ x).item()
                
            ucb_scores.append(mean + beta_t * math.sqrt(max(var, 0)))
            
        return int(np.argmax(ucb_scores))
    
    def propose_new_x(self, direction):
        """選択された方向に沿った最適化"""
        ei = ExpectedImprovement(self.model, best_f=self.best_value, maximize=False)
        
        # 方向に沿った探索範囲の計算
        active_dims_mask = direction.abs() > 1e-9
        if not active_dims_mask.any():
            lb, ub = -1.0, 1.0
        else:
            ratios_lower = (self.bounds[0] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            ratios_upper = (self.bounds[1] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            
            t_bounds = torch.zeros(self.dim, 2, device=self.X.device)
            t_bounds[:, 0] = torch.minimum(ratios_lower, ratios_upper)
            t_bounds[:, 1] = torch.maximum(ratios_lower, ratios_upper)
            
            lb = -float('inf')
            ub = float('inf')
            for i in range(self.dim):
                if active_dims_mask[i]:
                    lb = max(lb, t_bounds[i, 0].item())
                    ub = min(ub, t_bounds[i, 1].item())
                    
        if lb > ub:
            domain_width = (self.bounds[1, 0] - self.bounds[0, 0]).item()
            lb = -0.1 * domain_width
            ub = 0.1 * domain_width
            
        one_d_bounds = torch.tensor([[lb], [ub]], dtype=torch.float32, device=self.X.device)
        
        def ei_on_line(t_scalar_tensor):
            t_values = t_scalar_tensor.squeeze(-1)
            points_on_line = self.best_point.unsqueeze(0) + t_values.reshape(-1, 1) * direction.unsqueeze(0)
            points_on_line_clamped = torch.clamp(points_on_line, self.bounds[0].unsqueeze(0), self.bounds[1].unsqueeze(0))
            return ei(points_on_line_clamped.unsqueeze(1))
        
        # 獲得関数の最適化
        cand_t, _ = optimize_acqf(
            ei_on_line,
            bounds=one_d_bounds,
            q=1,
            num_restarts=10,
            raw_samples=100
        )
        
        alpha_star = cand_t.item()
        new_x = self.best_point + alpha_star * direction
        new_x_clamped = torch.clamp(new_x, self.bounds[0], self.bounds[1])
        
        return new_x_clamped
    
    def optimize(self):
        """メインの最適化ループ"""
        # 初期化
        self.initialize()
        n_iter = self.n_initial
        
        while n_iter < self.n_max:
            self.total_iterations += 1
            
            # 探索方向の候補生成
            arms_features = self.generate_arms()
            
            # Linear UCBによる方向選択
            sel_idx = self.select_arm(arms_features)
            direction = arms_features[sel_idx]
            
            # 方向を記録
            self.direction_history.append(direction.clone().detach().cpu().numpy())
            
            # 選択された方向に沿った最適化
            new_x = self.propose_new_x(direction)
            
            # 予測と実際の評価
            with torch.no_grad():
                predicted_mean = self.model.posterior(new_x.unsqueeze(0)).mean.squeeze().item()
            actual_y = self.objective_function(new_x.unsqueeze(0)).squeeze().item()
            
            # 報酬の計算（ベースライン、L下限付きスケーリング）
            new_x_for_grad = new_x.clone().unsqueeze(0)
            new_x_for_grad.requires_grad_(True)
            
            # GPモデルで事後分布を取得
            posterior = self.model.posterior(new_x_for_grad)
            mean_at_new_x = posterior.mean
            
            # 勾配を計算
            mean_at_new_x.sum().backward()
            grad_vector = new_x_for_grad.grad.squeeze(0)
            
            # 報酬ベクトル（勾配の絶対値）
            reward_vector = grad_vector.abs()
            # L_hat更新（増加のみ）と下限定義
            grad_norm = reward_vector.norm().item()
            if grad_norm > self.L_hat:
                self.L_hat = grad_norm
            L_effective = max(self.L_hat, self.L_min)
            # 下限付きスケーリング
            reward_vector = reward_vector / L_effective
            
            # 報酬を記録
            self.reward_history.append(reward_vector.clone().detach().cpu().numpy())
            
            # Linear Banditパラメータの更新
            x_arm = direction.view(-1, 1)
            self.A += x_arm @ x_arm.t()
            self.b += reward_vector
            
            # データとモデルの更新
            self.X = torch.cat([self.X, new_x.unsqueeze(0)], 0)
            self.Y = torch.cat([self.Y, torch.tensor([[actual_y]], dtype=torch.float32, device=self.X.device)], 0)
            self.update_model()
            
            # 最良点の更新
            with torch.no_grad():
                posterior_mean = self.model.posterior(self.X).mean.squeeze(-1)
            current_best_idx = posterior_mean.argmin()
            self.best_value = posterior_mean[current_best_idx].item()
            self.best_point = self.X[current_best_idx]
            
            self.eval_history.append(self.best_value)
            n_iter += 1
                
        return self.best_point, self.best_value

print("ベースラインLinBandit-BOクラスの定義完了")
```

```python
# 長さスケール下限LinBandit-BO
class FloorLinBanditBO:
    """
    長さスケール下限LinBandit-BO: 入力正規化 + RBFの長さスケールに下限を課す
    """
    
    def __init__(self, objective_function, bounds, n_initial=5, n_max=100, 
                 coordinate_ratio=0.8, n_arms=None):
        self.objective_function = objective_function
        self.bounds = bounds.float()
        self.dim = bounds.shape[1]
        self.n_initial = n_initial
        self.n_max = n_max
        self.coordinate_ratio = coordinate_ratio
        
        # 0.5x arms設定
        self.n_arms = n_arms if n_arms is not None else max(1, self.dim // 2)
        
        # Linear Banditのパラメータ
        self.A = torch.eye(self.dim)
        self.b = torch.zeros(self.dim)
        
        # 初期点の生成
        self.X = torch.rand(n_initial, self.dim) * (bounds[1] - bounds[0]) + bounds[0]
        self.X = self.X.float()
        
        # 状態変数
        self.Y = None
        self.best_value = None
        self.best_point = None
        self.model = None
        self.eval_history = []
        self.theta_history = []
        self.scale_init = 1.0
        self.total_iterations = 0
        
        # 報酬履歴の記録
        self.reward_history = []
        self.lengthscale_history = []
        self.direction_history = []
        self.gradient_history = []
    def update_model(self):
        """ガウス過程モデルの更新（正規化+長さスケール下限）"""
        X_gp = torch.clamp((self.X - self.bounds[0]) / (self.bounds[1]-self.bounds[0]), 0.0, 1.0)
        base = RBFKernel(ard_num_dims=self.X.shape[-1], lengthscale_constraint=GreaterThan(0.05))
        kernel = ScaleKernel(base).to(X_gp)
        self.model = SingleTaskGP(X_gp, self.Y, covar_module=kernel)
        mll = ExactMarginalLogLikelihood(self.model.likelihood, self.model)
        fit_gpytorch_model(mll)


    def initialize(self):
        """初期化：初期点での評価とモデル構築"""
        y_val = self.objective_function(self.X)
        self.Y = y_val.unsqueeze(-1).float()
        
        # スケーリング係数の計算
        y_max, y_min = self.Y.max().item(), self.Y.min().item()
        self.scale_init = (y_max - y_min) if (y_max - y_min) != 0 else 1.0
        
        # モデルの初期化
        self.update_model()
        
        # 最良点の初期化
        Xn = torch.clamp((self.X - self.bounds[0])/(self.bounds[1]-self.bounds[0]),0.0,1.0)
        post_mean = self.model.posterior(Xn).mean.squeeze(-1)
        bi = post_mean.argmin()
        self.best_value = post_mean[bi].item()
        self.best_point = self.X[bi]
        self.eval_history = [self.best_value] * self.n_initial
        
    def generate_arms(self):
        """アーム生成（0.5x arms）"""
        num_coord = int(self.coordinate_ratio * self.n_arms)
        num_coord = min(num_coord, self.dim)
        
        # ランダムに座標を選択
        idxs = np.random.choice(self.dim, num_coord, replace=False)
        
        # 座標方向の生成
        coords = []
        for i in idxs:
            e = torch.zeros(self.dim, device=self.X.device)
            e[i] = 1.0
            coords.append(e)
            
        coord_arms = torch.stack(coords, 0) if coords else torch.zeros(0, self.dim, device=self.X.device)
        
        # ランダム方向の生成
        num_rand = self.n_arms - num_coord
        rand_arms = torch.randn(num_rand, self.dim, device=self.X.device) if num_rand > 0 else torch.zeros(0, self.dim, device=self.X.device)
        
        if num_rand > 0:
            norms = rand_arms.norm(dim=1, keepdim=True)
            rand_arms = torch.where(norms > 1e-9, rand_arms / norms, 
                                   torch.randn_like(rand_arms) / (torch.randn_like(rand_arms).norm(dim=1,keepdim=True)+1e-9))
            
        return torch.cat([coord_arms, rand_arms], 0)
    
    def select_arm(self, arms_features):
        """Linear UCBによる方向選択"""
        # LinUCBパラメータ
        sigma = 1.0
        L = 1.0
        lambda_reg = 1.0
        delta = 0.1
        S = 1.0
        
        # 現在のパラメータ推定
        A_inv = torch.inverse(self.A)
        theta = A_inv @ self.b
        self.theta_history.append(theta.clone())
        
        # 信頼幅の計算
        current_round_t = max(1, self.total_iterations)
        log_term_numerator = max(1e-9, 1 + (current_round_t - 1) * L**2 / lambda_reg)
        beta_t = (sigma * math.sqrt(self.dim * math.log(log_term_numerator / delta)) + 
                  math.sqrt(lambda_reg) * S)
        
        # UCBスコアの計算
        ucb_scores = []
        for i in range(arms_features.shape[0]):
            x = arms_features[i].view(-1, 1)
            mean = (theta.view(1, -1) @ x).item()
            try:
                var = (x.t() @ A_inv @ x).item()
            except torch.linalg.LinAlgError:
                var = (x.t() @ torch.linalg.pinv(self.A) @ x).item()
                
            ucb_scores.append(mean + beta_t * math.sqrt(max(var, 0)))
            
        return int(np.argmax(ucb_scores))
    
    def propose_new_x(self, direction):
        """選択された方向に沿った最適化"""
        ei = ExpectedImprovement(self.model, best_f=self.best_value, maximize=False)
        
        # 方向に沿った探索範囲の計算
        active_dims_mask = direction.abs() > 1e-9
        if not active_dims_mask.any():
            lb, ub = -1.0, 1.0
        else:
            ratios_lower = (self.bounds[0] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            ratios_upper = (self.bounds[1] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            
            t_bounds = torch.zeros(self.dim, 2, device=self.X.device)
            t_bounds[:, 0] = torch.minimum(ratios_lower, ratios_upper)
            t_bounds[:, 1] = torch.maximum(ratios_lower, ratios_upper)
            
            lb = -float('inf')
            ub = float('inf')
            for i in range(self.dim):
                if active_dims_mask[i]:
                    lb = max(lb, t_bounds[i, 0].item())
                    ub = min(ub, t_bounds[i, 1].item())
                    
        if lb > ub:
            domain_width = (self.bounds[1, 0] - self.bounds[0, 0]).item()
            lb = -0.1 * domain_width
            ub = 0.1 * domain_width
            
        one_d_bounds = torch.tensor([[lb], [ub]], dtype=torch.float32, device=self.X.device)
        
        def ei_on_line(t_scalar_tensor):
            t_values = t_scalar_tensor.squeeze(-1)
            points_on_line = self.best_point.unsqueeze(0) + t_values.reshape(-1, 1) * direction.unsqueeze(0)
            points_on_line_clamped = torch.clamp(points_on_line, self.bounds[0].unsqueeze(0), self.bounds[1].unsqueeze(0))
            pts_n = torch.clamp((points_on_line_clamped - self.bounds[0])/(self.bounds[1]-self.bounds[0]),0.0,1.0)
            return ei(pts_n.unsqueeze(1))
        
        # 獲得関数の最適化
        cand_t, _ = optimize_acqf(
            ei_on_line,
            bounds=one_d_bounds,
            q=1,
            num_restarts=10,
            raw_samples=100
        )
        
        alpha_star = cand_t.item()
        new_x = self.best_point + alpha_star * direction
        new_x_clamped = torch.clamp(new_x, self.bounds[0], self.bounds[1])
        
        return new_x_clamped
    
    def optimize(self):
        """メインの最適化ループ"""
        # 初期化
        self.initialize()
        n_iter = self.n_initial
        
        while n_iter < self.n_max:
            self.total_iterations += 1
            
            # 探索方向の候補生成
            arms_features = self.generate_arms()
            
            # Linear UCBによる方向選択
            sel_idx = self.select_arm(arms_features)
            direction = arms_features[sel_idx]
            
            # 方向を記録
            self.direction_history.append(direction.clone().detach().cpu().numpy())
            
            # 選択された方向に沿った最適化
            new_x = self.propose_new_x(direction)
            
            # 予測と実際の評価
            with torch.no_grad():
                xn = torch.clamp((new_x.unsqueeze(0) - self.bounds[0])/(self.bounds[1]-self.bounds[0]),0.0,1.0)
                predicted_mean = self.model.posterior(xn).mean.squeeze().item()
            actual_y = self.objective_function(new_x.unsqueeze(0)).squeeze().item()
            
            # 勾配の計算（正規化空間 -> 元空間へ変換）
            new_x_for_grad = new_x.clone().unsqueeze(0)
            new_x_for_grad.requires_grad_(True)
            x_norm = torch.clamp((new_x_for_grad - self.bounds[0])/(self.bounds[1]-self.bounds[0]), 0.0, 1.0)
            x_norm.retain_grad()
            posterior = self.model.posterior(x_norm)
            mean_at_new_x = posterior.mean
            mean_at_new_x.sum().backward()
            grad_normed = x_norm.grad.squeeze(0)
            scale = (self.bounds[1]-self.bounds[0])
            grad_vector = grad_normed / (scale + 1e-12)
            reward_vector = grad_vector.abs()
            # L_hat更新と下限スケーリング
            grad_norm = reward_vector.norm().item()
            if not hasattr(self, 'L_min'):
                self.L_min = 0.1
                self.L_hat = max(1.0, self.L_min)
            if grad_norm > self.L_hat:
                self.L_hat = grad_norm
            L_effective = max(self.L_hat, self.L_min)
            reward_vector = reward_vector / L_effective
            # 長さスケールを記録
            try:
                ls = self.model.covar_module.base_kernel.lengthscale.detach().cpu().numpy().reshape(-1)
                if not hasattr(self, 'lengthscale_history'):
                    self.lengthscale_history = []
                self.lengthscale_history.append(ls.tolist())
            except Exception:
                pass
            # 報酬を記録
            self.reward_history.append(reward_vector.clone().detach().cpu().numpy())
            
            # Linear Banditパラメータの更新
            x_arm = direction.view(-1, 1)
            self.A += x_arm @ x_arm.t()
            self.b += reward_vector
            
            # データとモデルの更新
            self.X = torch.cat([self.X, new_x.unsqueeze(0)], 0)
            self.Y = torch.cat([self.Y, torch.tensor([[actual_y]], dtype=torch.float32, device=self.X.device)], 0)
            self.update_model()
            
            # 最良点の更新
            with torch.no_grad():
                Xn2 = torch.clamp((self.X - self.bounds[0])/(self.bounds[1]-self.bounds[0]),0.0,1.0)
                posterior_mean = self.model.posterior(Xn2).mean.squeeze(-1)
            current_best_idx = posterior_mean.argmin()
            self.best_value = posterior_mean[current_best_idx].item()
            self.best_point = self.X[current_best_idx]
            
            self.eval_history.append(self.best_value)
            n_iter += 1
                
        return self.best_point, self.best_value

print("長さスケール下限LinBandit-BOクラスの定義完了")
```

```python
# テスト関数の定義
def styblinski_tang_effective(x, effective_dims=5):
    """Styblinski-Tang関数（有効次元のみ）"""
    if not torch.is_tensor(x):
        x = torch.tensor(x, dtype=torch.float32)
    
    # 有効次元のみを使用
    x_eff = x[..., :effective_dims]
    return 0.5 * torch.sum(x_eff**4 - 16.0*x_eff**2 + 5.0*x_eff, dim=-1)

def rastrigin_effective(x, effective_dims=5):
    """Rastrigin関数（有効次元のみ）"""
    if not torch.is_tensor(x):
        x = torch.tensor(x, dtype=torch.float32)
    
    # 有効次元のみを使用
    x_eff = x[..., :effective_dims]
    return torch.sum(x_eff**2 - 10.0*torch.cos(2*math.pi*x_eff) + 10.0, dim=-1)

def ackley_effective(x, effective_dims=5):
    """Ackley関数（有効次元のみ）"""
    if not torch.is_tensor(x):
        x = torch.tensor(x, dtype=torch.float32)
    
    # 有効次元のみを使用
    x_eff = x[..., :effective_dims]
    d = x_eff.shape[-1]
    
    sum1 = torch.sum(x_eff**2, dim=-1)
    sum2 = torch.sum(torch.cos(2*math.pi*x_eff), dim=-1)
    
    return -20.0 * torch.exp(-0.2 * torch.sqrt(sum1/d)) - torch.exp(sum2/d) + 20.0 + math.e

# テスト関数の設定
test_functions = {
    'Styblinski-Tang': styblinski_tang_effective,
    'Rastrigin': rastrigin_effective,
    'Ackley': ackley_effective
}

# 大域的最適値
global_optima = {
    'Styblinski-Tang': -39.16599 * 5,  # 5次元
    'Rastrigin': 0.0,
    'Ackley': 0.0
}

print("テスト関数の定義完了")
```

```python
# 実験実行関数
def run_single_experiment(algorithm_class, objective_function, bounds, algorithm_name):
    """単一実験の実行"""
    optimizer = algorithm_class(
        objective_function=objective_function,
        bounds=bounds,
        n_initial=5,
        n_max=300,
        coordinate_ratio=0.8
    )
    
    # 同じ初期点を使用するため、シードを固定
    torch.manual_seed(42)
    np.random.seed(42)
    
    optimizer.optimize()
    
    result = {
        'eval_history': optimizer.eval_history,
        'best_value': optimizer.best_value,
        'theta_history': optimizer.theta_history,
        'reward_history': optimizer.reward_history,
        'direction_history': optimizer.direction_history
    }
    
    # 追加履歴（必要なら）
    if hasattr(optimizer, 'lengthscale_history'):
        result['lengthscale_history'] = optimizer.lengthscale_history
        result['gradient_history'] = getattr(optimizer, 'gradient_history', None)
    
    return result

def run_comparison_experiment(func_name, objective_function, n_runs=20):
    """比較実験の実行"""
    print(f"\n=== {func_name} 実験開始 ===")
    
    dim = 20
    bounds = torch.tensor([[-5.0]*dim, [5.0]*dim], dtype=torch.float32)
    
    results = {
        'Baseline': [],
        'Floor': []
    }
    
    algorithms = {
        'Baseline': BaselineLinBanditBO,
        'Floor': FloorLinBanditBO
    }
    
    for alg_name, alg_class in algorithms.items():
        print(f"{alg_name}の実験中...")
        for run_idx in range(n_runs):
            print(f"\r  実行中: {run_idx + 1}/{n_runs}", end="", flush=True)
            
            # 各実行で異なるシードを使用
            torch.manual_seed(run_idx * 100)
            np.random.seed(run_idx * 100)
            
            result = run_single_experiment(alg_class, objective_function, bounds, alg_name)
            results[alg_name].append(result)
        
        print(f"\n  {alg_name}完了")
    
    return results

print("実験実行関数の定義完了")
```

```python
# 追加計測クラス（L_hat/|∇μ|/r≈1 を追跡する実装）
HIT_THRESHOLD = 0.95  # r≈1 の判定しきい値

class InstrumentedBaselineLinBanditBO:
    def __init__(self, objective_function, bounds, n_initial=5, n_max=100,
                 coordinate_ratio=0.8, n_arms=None):
        self.objective_function = objective_function
        self.bounds = bounds.float()
        self.dim = bounds.shape[1]
        self.n_initial = n_initial
        self.n_max = n_max
        self.coordinate_ratio = coordinate_ratio
        self.n_arms = n_arms if n_arms is not None else max(1, self.dim // 2)
        self.A = torch.eye(self.dim)
        self.b = torch.zeros(self.dim)
        self.X = torch.rand(n_initial, self.dim) * (bounds[1] - bounds[0]) + bounds[0]
        self.X = self.X.float()
        self.Y = None
        self.best_value = None
        self.best_point = None
        self.model = None
        self.eval_history = []
        self.theta_history = []
        self.scale_init = 1.0
        self.total_iterations = 0
        # 履歴
        self.reward_history = []            # スケーリング後 r_i
        self.direction_history = []
        self.grad_norm_history = []        # 正規化前 ∥∇μ∥
        self.L_hat_history = []            # 推定リプシッツ L_hat の推移
        self.r_one_hit_ratio_history = []  # r≈1 の到達率（各反復の次元平均）
        # スパイク抑制用スケーリングの下限
        self.L_min = 0.1
        self.L_hat = max(1.0, self.L_min)

    def update_model(self):
        base = RBFKernel(ard_num_dims=self.X.shape[-1])
        kernel = ScaleKernel(base).to(self.X)
        self.model = SingleTaskGP(self.X, self.Y, covar_module=kernel)
        mll = ExactMarginalLogLikelihood(self.model.likelihood, self.model)
        fit_gpytorch_model(mll)

    def initialize(self):
        y_val = self.objective_function(self.X)
        self.Y = y_val.unsqueeze(-1).float()
        y_max, y_min = self.Y.max().item(), self.Y.min().item()
        self.scale_init = (y_max - y_min) if (y_max - y_min) != 0 else 1.0
        self.update_model()
        post_mean = self.model.posterior(self.X).mean.squeeze(-1)
        bi = post_mean.argmin()
        self.best_value = post_mean[bi].item()
        self.best_point = self.X[bi]
        self.eval_history = [self.best_value] * self.n_initial

    def generate_arms(self):
        num_coord = int(self.coordinate_ratio * self.n_arms)
        num_coord = min(num_coord, self.dim)
        idxs = np.random.choice(self.dim, num_coord, replace=False)
        coords = []
        for i in idxs:
            e = torch.zeros(self.dim, device=self.X.device)
            e[i] = 1.0
            coords.append(e)
        coord_arms = torch.stack(coords, 0) if coords else torch.zeros(0, self.dim, device=self.X.device)
        num_rand = self.n_arms - num_coord
        rand_arms = torch.randn(num_rand, self.dim, device=self.X.device) if num_rand > 0 else torch.zeros(0, self.dim, device=self.X.device)
        if num_rand > 0:
            norms = rand_arms.norm(dim=1, keepdim=True)
            rand_arms = torch.where(norms > 1e-9, rand_arms / norms,
                                   torch.randn_like(rand_arms) / (torch.randn_like(rand_arms).norm(dim=1, keepdim=True)+1e-9))
        return torch.cat([coord_arms, rand_arms], 0)

    def select_arm(self, arms_features):
        sigma = 1.0; L = 1.0; lambda_reg = 1.0; delta = 0.1; S = 1.0
        A_inv = torch.inverse(self.A)
        theta = A_inv @ self.b
        self.theta_history.append(theta.clone())
        current_round_t = max(1, self.total_iterations)
        log_term_numerator = max(1e-9, 1 + (current_round_t - 1) * L**2 / lambda_reg)
        beta_t = (sigma * math.sqrt(self.dim * math.log(log_term_numerator / delta)) + math.sqrt(lambda_reg) * S)
        ucb_scores = []
        for i in range(arms_features.shape[0]):
            x = arms_features[i].view(-1, 1)
            mean = (theta.view(1, -1) @ x).item()
            try:
                var = (x.t() @ A_inv @ x).item()
            except torch.linalg.LinAlgError:
                var = (x.t() @ torch.linalg.pinv(self.A) @ x).item()
            ucb_scores.append(mean + beta_t * math.sqrt(max(var, 0)))
        return int(np.argmax(ucb_scores))

    def propose_new_x(self, direction):
        ei = ExpectedImprovement(self.model, best_f=self.best_value, maximize=False)
        active_dims_mask = direction.abs() > 1e-9
        if not active_dims_mask.any():
            lb, ub = -1.0, 1.0
        else:
            ratios_lower = (self.bounds[0] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            ratios_upper = (self.bounds[1] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            t_bounds = torch.zeros(self.dim, 2, device=self.X.device)
            t_bounds[:, 0] = torch.minimum(ratios_lower, ratios_upper)
            t_bounds[:, 1] = torch.maximum(ratios_lower, ratios_upper)
            lb = -float('inf'); ub = float('inf')
            for i in range(self.dim):
                if active_dims_mask[i]:
                    lb = max(lb, t_bounds[i, 0].item())
                    ub = min(ub, t_bounds[i, 1].item())
        if lb > ub:
            domain_width = (self.bounds[1, 0] - self.bounds[0, 0]).item()
            lb = -0.1 * domain_width; ub = 0.1 * domain_width
        one_d_bounds = torch.tensor([[lb], [ub]], dtype=torch.float32, device=self.X.device)
        def ei_on_line(t_scalar_tensor):
            t_values = t_scalar_tensor.squeeze(-1)
            points_on_line = self.best_point.unsqueeze(0) + t_values.reshape(-1, 1) * direction.unsqueeze(0)
            points_on_line_clamped = torch.clamp(points_on_line, self.bounds[0].unsqueeze(0), self.bounds[1].unsqueeze(0))
            return ei(points_on_line_clamped.unsqueeze(1))
        cand_t, _ = optimize_acqf(ei_on_line, bounds=one_d_bounds, q=1, num_restarts=10, raw_samples=100)
        alpha_star = cand_t.item()
        new_x = self.best_point + alpha_star * direction
        return torch.clamp(new_x, self.bounds[0], self.bounds[1])

    def optimize(self):
        self.initialize()
        n_iter = self.n_initial
        while n_iter < self.n_max:
            self.total_iterations += 1
            arms_features = self.generate_arms()
            sel_idx = self.select_arm(arms_features)
            direction = arms_features[sel_idx]
            new_x = self.propose_new_x(direction)
            # 予測と実際の評価（ログ用途のみ）
            with torch.no_grad():
                predicted_mean = self.model.posterior(new_x.unsqueeze(0)).mean.squeeze().item()
            actual_y = self.objective_function(new_x.unsqueeze(0)).squeeze().item()
            # 勾配と報酬ベクトル
            new_x_for_grad = new_x.clone().unsqueeze(0)
            new_x_for_grad.requires_grad_(True)
            posterior = self.model.posterior(new_x_for_grad)
            posterior.mean.sum().backward()
            grad_vector = new_x_for_grad.grad.squeeze(0)
            reward_vector = grad_vector.abs()
            # L_hat 更新と履歴
            grad_norm = reward_vector.norm().item()
            if grad_norm > self.L_hat:
                self.L_hat = grad_norm
            L_effective = max(self.L_hat, self.L_min)
            self.grad_norm_history.append(grad_norm)
            self.L_hat_history.append(self.L_hat)
            scaled_reward_vector = reward_vector / L_effective
            hit_ratio = float((scaled_reward_vector >= HIT_THRESHOLD).float().mean().item())
            self.r_one_hit_ratio_history.append(hit_ratio)
            # Linear Bandit の更新
            x_arm = direction.view(-1, 1)
            self.A += x_arm @ x_arm.t()
            self.b += scaled_reward_vector
            self.direction_history.append(direction.detach().clone())
            self.reward_history.append(scaled_reward_vector.detach().clone())
            # データ・モデル更新
            self.X = torch.cat([self.X, new_x.unsqueeze(0)], 0)
            self.Y = torch.cat([self.Y, torch.tensor([[actual_y]], dtype=torch.float32, device=self.X.device)], 0)
            self.update_model()
            # 最良点更新
            with torch.no_grad():
                posterior_mean = self.model.posterior(self.X).mean.squeeze(-1)
            bi = posterior_mean.argmin()
            self.best_value = posterior_mean[bi].item()
            self.best_point = self.X[bi]
            self.eval_history.append(self.best_value)
            n_iter += 1
        return self.best_point, self.best_value

class InstrumentedFloorLinBanditBO(InstrumentedBaselineLinBanditBO):
    def __init__(self, objective_function, bounds, n_initial=5, n_max=100,
                 coordinate_ratio=0.8, n_arms=None):
        super().__init__(objective_function, bounds, n_initial, n_max, coordinate_ratio, n_arms)
        self.lengthscale_history = []
        self._range = (self.bounds[1] - self.bounds[0]).float()

    def _to_normalized(self, X):
        return torch.clamp((X - self.bounds[0]) / self._range, 0.0, 1.0)

    def update_model(self):
        X_gp = self._to_normalized(self.X)
        base = RBFKernel(ard_num_dims=self.X.shape[-1], lengthscale_constraint=GreaterThan(0.05))
        kernel = ScaleKernel(base).to(X_gp)
        self.model = SingleTaskGP(X_gp, self.Y, covar_module=kernel)
        mll = ExactMarginalLogLikelihood(self.model.likelihood, self.model)
        fit_gpytorch_model(mll)
        # 学習後の長さスケールを記録
        try:
            ls = self.model.covar_module.base_kernel.lengthscale.detach().cpu().numpy().tolist()
            self.lengthscale_history.append(ls)
        except Exception:
            pass

    def initialize(self):
        y_val = self.objective_function(self.X)
        self.Y = y_val.unsqueeze(-1).float()
        y_max, y_min = self.Y.max().item(), self.Y.min().item()
        self.scale_init = (y_max - y_min) if (y_max - y_min) != 0 else 1.0
        self.update_model()
        Xn = self._to_normalized(self.X)
        post_mean = self.model.posterior(Xn).mean.squeeze(-1)
        bi = post_mean.argmin()
        self.best_value = post_mean[bi].item()
        self.best_point = self.X[bi]
        self.eval_history = [self.best_value] * self.n_initial

    def propose_new_x(self, direction):
        ei = ExpectedImprovement(self.model, best_f=self.best_value, maximize=False)
        active_dims_mask = direction.abs() > 1e-9
        if not active_dims_mask.any():
            lb, ub = -1.0, 1.0
        else:
            ratios_lower = (self.bounds[0] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            ratios_upper = (self.bounds[1] - self.best_point) / (direction + 1e-12 * (~active_dims_mask))
            t_bounds = torch.zeros(self.dim, 2, device=self.X.device)
            t_bounds[:, 0] = torch.minimum(ratios_lower, ratios_upper)
            t_bounds[:, 1] = torch.maximum(ratios_lower, ratios_upper)
            lb = -float('inf'); ub = float('inf')
            for i in range(self.dim):
                if active_dims_mask[i]:
                    lb = max(lb, t_bounds[i, 0].item())
                    ub = min(ub, t_bounds[i, 1].item())
        if lb > ub:
            domain_width = (self.bounds[1, 0] - self.bounds[0, 0]).item()
            lb = -0.1 * domain_width; ub = 0.1 * domain_width
        one_d_bounds = torch.tensor([[lb], [ub]], dtype=torch.float32, device=self.X.device)
        def ei_on_line(t_scalar_tensor):
            t_values = t_scalar_tensor.squeeze(-1)
            points_on_line = self.best_point.unsqueeze(0) + t_values.reshape(-1, 1) * direction.unsqueeze(0)
            points_on_line_clamped = torch.clamp(points_on_line, self.bounds[0].unsqueeze(0), self.bounds[1].unsqueeze(0))
            pts = self._to_normalized(points_on_line_clamped)
            return ei(pts.unsqueeze(1))
        cand_t, _ = optimize_acqf(ei_on_line, bounds=one_d_bounds, q=1, num_restarts=10, raw_samples=100)
        alpha_star = cand_t.item()
        new_x = self.best_point + alpha_star * direction
        return torch.clamp(new_x, self.bounds[0], self.bounds[1])

    def optimize(self):
        self.initialize()
        n_iter = self.n_initial
        while n_iter < self.n_max:
            self.total_iterations += 1
            arms_features = self.generate_arms()
            sel_idx = self.select_arm(arms_features)
            direction = arms_features[sel_idx]
            new_x = self.propose_new_x(direction)
            with torch.no_grad():
                x_q = self._to_normalized(new_x.unsqueeze(0))
                _ = self.model.posterior(x_q).mean.squeeze().item()
            actual_y = self.objective_function(new_x.unsqueeze(0)).squeeze().item()
            # 勾配（実座標系）と報酬ベクトル
            new_x_for_grad = self._to_normalized(new_x.clone()).unsqueeze(0)
            new_x_for_grad.requires_grad_(True)
            posterior = self.model.posterior(new_x_for_grad)
            posterior.mean.sum().backward()
            grad_vector_normed = new_x_for_grad.grad.squeeze(0)
            grad_vector = grad_vector_normed / (self._range + 1e-12)
            reward_vector = grad_vector.abs()
            grad_norm = reward_vector.norm().item()
            if grad_norm > self.L_hat:
                self.L_hat = grad_norm
            L_effective = max(self.L_hat, self.L_min)
            self.grad_norm_history.append(grad_norm)
            self.L_hat_history.append(self.L_hat)
            scaled_reward_vector = reward_vector / L_effective
            hit_ratio = float((scaled_reward_vector >= HIT_THRESHOLD).float().mean().item())
            self.r_one_hit_ratio_history.append(hit_ratio)
            # Linear Bandit の更新
            x_arm = direction.view(-1, 1)
            self.A += x_arm @ x_arm.t()
            self.b += scaled_reward_vector
            self.direction_history.append(direction.detach().clone())
            self.reward_history.append(scaled_reward_vector.detach().clone())
            # データ・モデル更新
            self.X = torch.cat([self.X, new_x.unsqueeze(0)], 0)
            self.Y = torch.cat([self.Y, torch.tensor([[actual_y]], dtype=torch.float32, device=self.X.device)], 0)
            self.update_model()
            # 最良点更新
            with torch.no_grad():
                X_query2 = self._to_normalized(self.X)
                posterior_mean = self.model.posterior(X_query2).mean.squeeze(-1)
            bi = posterior_mean.argmin()
            self.best_value = posterior_mean[bi].item()
            self.best_point = self.X[bi]
            self.eval_history.append(self.best_value)
            n_iter += 1
        return self.best_point, self.best_value
```

```python
# 実験関数（計測対応版に上書き）
def run_single_experiment(algorithm_class, objective_function, bounds, algorithm_name):
    optimizer = algorithm_class(
        objective_function=objective_function,
        bounds=bounds,
        n_initial=5,
        n_max=300,
        coordinate_ratio=0.8
    )
    # 同一初期点のためのシード固定
    torch.manual_seed(42)
    np.random.seed(42)
    optimizer.optimize()
    result = {
        'eval_history': optimizer.eval_history,
        'best_value': optimizer.best_value,
        'theta_history': optimizer.theta_history,
        'reward_history': optimizer.reward_history,
        'direction_history': optimizer.direction_history
    }
    # 追加履歴
    if hasattr(optimizer, 'lengthscale_history'):
        result['lengthscale_history'] = optimizer.lengthscale_history
    if hasattr(optimizer, 'gradient_history') and optimizer.gradient_history:
        result['gradient_history'] = optimizer.gradient_history
    if hasattr(optimizer, 'L_hat_history') and optimizer.L_hat_history:
        result['L_hat_history'] = optimizer.L_hat_history
    if hasattr(optimizer, 'grad_norm_history') and optimizer.grad_norm_history:
        result['grad_norm_history'] = optimizer.grad_norm_history
    if hasattr(optimizer, 'r_one_hit_ratio_history') and optimizer.r_one_hit_ratio_history:
        result['r_one_hit_ratio_history'] = optimizer.r_one_hit_ratio_history
    if hasattr(optimizer, 'L_hat'):
        result['L_hat_final'] = optimizer.L_hat
    return result

def run_comparison_experiment(func_name, objective_function, n_runs=20):
    print(f"\n=== {func_name} 実験開始 ===")
    dim = 20
    bounds = torch.tensor([[-5.0]*dim, [5.0]*dim], dtype=torch.float32)
    results = { 'Baseline': [], 'Floor': [] }
    algorithms = {
        'Baseline': InstrumentedBaselineLinBanditBO,
        'Floor': InstrumentedFloorLinBanditBO
    }
    for alg_name, alg_class in algorithms.items():
        print(f"{alg_name}の実験中...")
        for run_idx in range(n_runs):
            print(f"\r  実行中: {run_idx + 1}/{n_runs}", end="", flush=True)
            torch.manual_seed(run_idx * 100)
            np.random.seed(run_idx * 100)
            result = run_single_experiment(alg_class, objective_function, bounds, alg_name)
            results[alg_name].append(result)
        print(f"\n  {alg_name}完了")
    return results
print("計測対応版の実験関数を定義しました")
```

```python
# 追加診断（L_hat推移/最終値, |∇μ|分布, r≈1 到達率）の集計と可視化
def analyze_additional_metrics(results_dict, func_name, hit_threshold=0.95):
    # L_hat, 勾配ノルム, r≈1 のデータフレームを作成
    rows_lhat, rows_grad, rows_hit = [], [], []
    for alg_name, results in results_dict.items():
        for run_idx, r in enumerate(results):
            # L_hat の推移
            if 'L_hat_history' in r and r['L_hat_history']:
                for it, val in enumerate(r['L_hat_history']):
                    rows_lhat.append({'Algorithm': alg_name, 'Run': run_idx, 'Iteration': it + 5, 'L_hat': float(val)})
            # 勾配ノルムの推移
            if 'grad_norm_history' in r and r['grad_norm_history']:
                for it, val in enumerate(r['grad_norm_history']):
                    rows_grad.append({'Algorithm': alg_name, 'Run': run_idx, 'Iteration': it + 5, 'GradNorm': float(val)})
            # r≈1 到達率
            if 'r_one_hit_ratio_history' in r and r['r_one_hit_ratio_history']:
                for it, val in enumerate(r['r_one_hit_ratio_history']):
                    rows_hit.append({'Algorithm': alg_name, 'Run': run_idx, 'Iteration': it + 5, 'HitRatio': float(val)})
            else:
                # reward_history から算出（互換用）
                if 'reward_history' in r and r['reward_history']:
                    for it, rewards in enumerate(r['reward_history']):
                        rewards_np = np.asarray(rewards, dtype=float)
                        ratio = float((rewards_np >= hit_threshold).mean())
                        rows_hit.append({'Algorithm': alg_name, 'Run': run_idx, 'Iteration': it + 5, 'HitRatio': ratio})
    df_lhat = pd.DataFrame(rows_lhat) if rows_lhat else pd.DataFrame(columns=['Algorithm','Run','Iteration','L_hat'])
    df_grad = pd.DataFrame(rows_grad) if rows_grad else pd.DataFrame(columns=['Algorithm','Run','Iteration','GradNorm'])
    df_hit  = pd.DataFrame(rows_hit)  if rows_hit  else pd.DataFrame(columns=['Algorithm','Run','Iteration','HitRatio'])
    # 可視化
    import matplotlib.pyplot as plt
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    # 1) L_hat 推移
    ax1 = axes[0,0]
    if not df_lhat.empty:
        for alg_name, color, label in [('Baseline','blue','ベースライン'), ('Floor','red','長さスケール下限')]:
            sub = df_lhat[df_lhat['Algorithm']==alg_name]
            if sub.empty: continue
            mean = sub.groupby('Iteration')['L_hat'].mean(); std = sub.groupby('Iteration')['L_hat'].std()
            x = mean.index.values; ax1.plot(x, mean.values, color=color, label=label, linewidth=2)
            ax1.fill_between(x, mean.values-std.values, mean.values+std.values, color=color, alpha=0.2)
        ax1.set_title(f'{func_name}: L_hat の推移（平均±STD）')
        ax1.set_xlabel('Iteration'); ax1.set_ylabel('L_hat'); ax1.grid(True, alpha=0.3); ax1.legend()
    # 2) 最終 L_hat の箱ひげ
    ax2 = axes[0,1]
    finals = []
    labels = []
    for alg in ['Baseline','Floor']:
        sub = df_lhat[df_lhat['Algorithm']==alg]
        if not sub.empty:
            last = sub.groupby(['Algorithm','Run']).apply(lambda d: d.sort_values('Iteration').iloc[-1]['L_hat']).values
            finals.append(last); labels.append('ベースライン' if alg=='Baseline' else '長さスケール下限')
    if finals:
        box = ax2.boxplot(finals, labels=labels, patch_artist=True)
        if len(finals)>=1: box['boxes'][0].set_facecolor('lightblue')
        if len(finals)>=2: box['boxes'][1].set_facecolor('lightcoral')
        ax2.set_title(f'{func_name}: 最終 L_hat 分布'); ax2.grid(True, alpha=0.3)
    # 3) 正規化前 ∥∇μ∥ 分布
    ax3 = axes[1,0]
    for alg_name, color, label in [('Baseline','blue','ベースライン'), ('Floor','red','長さスケール下限')]:
        sub = df_grad[df_grad['Algorithm']==alg_name]
        if sub.empty: continue
        ax3.hist(sub['GradNorm'].values, bins=50, alpha=0.5, color=color, label=label, edgecolor='black')
    ax3.set_title(f'{func_name}: 正規化前の勾配ノルム分布 (|∇μ|)'); ax3.set_xlabel('∥∇μ∥'); ax3.set_ylabel('Frequency'); ax3.legend(); ax3.grid(True, alpha=0.3)
    # 4) r≈1 到達率の推移
    ax4 = axes[1,1]
    if not df_hit.empty:
        for alg_name, color, label in [('Baseline','blue','ベースライン'), ('Floor','red','長さスケール下限')]:
            sub = df_hit[df_hit['Algorithm']==alg_name]
            if sub.empty: continue
            mean = sub.groupby('Iteration')['HitRatio'].mean(); x = mean.index.values
            ax4.plot(x, mean.values, color=color, label=label, linewidth=2)
        ax4.set_title(f'{func_name}: 報酬の上限到達率 (r≈1)'); ax4.set_xlabel('Iteration'); ax4.set_ylabel('Hit Ratio'); ax4.grid(True, alpha=0.3); ax4.legend()
    plt.tight_layout()
    plt.savefig(f'{output_dir}/{func_name}_diagnostics.png', dpi=300, bbox_inches='tight')
    plt.show()
    # CSV 保存
    if not df_lhat.empty: df_lhat.to_csv(f'{output_dir}/{func_name}_lhat_history.csv', index=False)
    if not df_grad.empty: df_grad.to_csv(f'{output_dir}/{func_name}_grad_norms.csv', index=False)
    if not df_hit.empty:  df_hit.to_csv(f'{output_dir}/{func_name}_r_upper_hit_rate.csv', index=False)
    return df_lhat, df_grad, df_hit
print("追加診断関数を定義しました")
```

```python
# 全体の比較サマリー
print("\n" + "="*70)
print("全体結果サマリー: ベースライン vs 長さスケール下限 LinBandit-BO")
print("="*70)

print("\n各テスト関数の詳細比較:")
for func_name in test_functions.keys():
    results = all_results[func_name]
    
    gradient_final = [result['best_value'] for result in results['Baseline']]
    cosine_final = [result['best_value'] for result in results['Floor']]
    
    # 改善率の計算
    gradient_mean = np.mean(gradient_final)
    cosine_mean = np.mean(cosine_final)
    improvement_rate = (gradient_mean - cosine_mean) / abs(gradient_mean) * 100
    
    print(f"\n{func_name}:")
    print(f"  ベースライン平均: {gradient_mean:.6f} (±{np.std(gradient_final):.6f})")
    print(f"  長さスケール下限平均: {cosine_mean:.6f} (±{np.std(cosine_final):.6f})")
    print(f"  改善率（コサイン類似度の観点から）: {improvement_rate:.2f}%")
    print(f"  グローバル最適値: {global_optima[func_name]:.2f}")
    
    # 最適値からの距離
    gradient_gap = gradient_mean - global_optima[func_name]
    cosine_gap = cosine_mean - global_optima[func_name]
    print(f"  ベースラインの最適値からの差: {gradient_gap:.6f}")
    print(f"  長さスケール下限の最適値からの差: {cosine_gap:.6f}")
    
    # 統計的有意差の検定（Mann-Whitney U test）
    from scipy.stats import mannwhitneyu
    statistic, p_value = mannwhitneyu(gradient_final, cosine_final, alternative='two-sided')
    print(f"  統計的有意差 (p-value): {p_value:.4f}")
    
    if p_value < 0.05:
        if gradient_mean < cosine_mean:
            print(f"  → ベースラインが統計的に有意に優秀")
        else:
            print(f"  → 長さスケール下限が統計的に有意に優秀")
    else:
        print(f"  → 統計的有意差なし")

# 勝率の計算
print(f"\n勝敗比較（各実行での優劣）:")
for func_name in test_functions.keys():
    results = all_results[func_name]
    gradient_final = [result['best_value'] for result in results['Baseline']]
    cosine_final = [result['best_value'] for result in results['Floor']]
    
    gradient_wins = sum(1 for g, c in zip(gradient_final, cosine_final) if g < c)
    ties = sum(1 for g, c in zip(gradient_final, cosine_final) if abs(g - c) < 1e-6)
    cosine_wins = n_runs - gradient_wins - ties
    
    print(f"  {func_name}: ベースライン{gradient_wins}勝, 長さスケール下限{cosine_wins}勝, 引き分け{ties}")

# コサイン類似度の統計
print(f"\nコサイン類似度の統計:")
for func_name in test_functions.keys():
    results = all_results[func_name]
    all_cosine_similarities = []
    for result in results['Floor']:
        if 'cosine_similarity_history' in result:
            all_cosine_similarities.extend(result['cosine_similarity_history'])
    
    if all_cosine_similarities:
        print(f"  {func_name}:")
        print(f"    平均: {np.mean(all_cosine_similarities):.4f}")
        print(f"    標準偏差: {np.std(all_cosine_similarities):.4f}")
        print(f"    最小値: {np.min(all_cosine_similarities):.4f}")
        print(f"    最大値: {np.max(all_cosine_similarities):.4f}")

print("\n" + "="*70)
print("実験設定:")
print("- ベースライン: 勾配の絶対値を報酬とし、L下限付きスケーリングを適用")
print("- 長さスケール下限: 勾配と選択方向のコサイン類似度を報酬として使用") 
print("- 両アルゴリズムとも: 0.5x arms、coordinate_ratio=0.8")
print("- 20次元問題（有効次元: 0-4）")
print("- 各アルゴリズム20回独立実行")
print("- 300回評価")
print("="*70)

print(f"\n出力ファイル:")
print(f"- 収束履歴プロット: {output_dir}/[function_name]_comparison.png")
print(f"- 報酬履歴分析: {output_dir}/[function_name]_reward_analysis.png")
print(f"- 報酬履歴CSV: {output_dir}/[function_name]_reward_history.csv")
print(f"- 次元別要約CSV: {output_dir}/[function_name]_dimension_summary.csv")
print(f"- 実験結果numpy: {output_dir}/[function_name]_results.npy")
```
